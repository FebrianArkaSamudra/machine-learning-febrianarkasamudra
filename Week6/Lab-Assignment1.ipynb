# Question Number 1

## Step 1 - Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.cluster.hierarchy as sch # Build dendogram
from sklearn.metrics import silhouette_score # Silhouette Coefficient
from sklearn.cluster import AgglomerativeClustering  # Build Cluster using Agglomerative Method
```

## Step 2 - Load Dataset

```python
df = pd.read_csv('CC_Customer_Dataset.csv')

df.info()
```

```python
df.head()
```

## Step 3 - Slicing the Data with All Possible Variables

```python
# Slicing - Select only specific columns (Annual Income and Spending Score)
X = df.iloc[:, [2, 3, 4, 5, 6]].values  # .values converts selected data into a NumPy array
```

## Step 4 - Visualize the Dendrogram to Determine the Cluster by each Threshold

```python
# Build a dendrogram using the Ward method
# (Ward minimizes variance between merged clusters)
dendrogram = sch.dendrogram(sch.linkage(X, method="ward"))

# Plot the dendrogram
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

based on the results it determine into 2 cluster which is, it is like broad segmentation where maybe if using 3 or 4 segmentation still make sense, but i'll try to create into 3 clusters

```python
# Build clusters using Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
y_hc = hc.fit_predict(X)
```

## Step 5 - Visualization of Clusters

```python
# Visualize The Result
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c="red", label="Cluster 1")
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c="blue", label="Cluster 2")
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c="green", label="Cluster 3")
plt.title("Clusters of customers")
plt.legend()
plt.show()
```

## Step 6 - Evaluation Cluster Results using Silhouette Score

```python
# Evaluate the clustering using silhouette coefficient
silhouette_avg = silhouette_score(X, y_hc)
print(f"Silhouette Coefficient: {silhouette_avg}")
```

# Question Number 2

## Step 1 - Slicing Data for Two Variables

```python
# Select only 'Avg_Credit_Limit' and 'Total_Credit_Cards' columns as features
X2 = df.loc[:, ["Avg_Credit_Limit", "Total_Credit_Cards"]].values
```

## Step 2 - Fit Agglomerative Clustering Model

```python
# Apply Agglomerative Clustering using the selected two features
hc2 = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
y_hc2 = hc2.fit_predict(X2)
```

## Step 3 - Visualize The Clusters with a Scatterplot 

```python
# Visualize the clusters using a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(X2[y_hc2 == 0, 0], X2[y_hc2 == 0, 1], s=100, c="red", label="Cluster 1")
plt.scatter(X2[y_hc2 == 1, 0], X2[y_hc2 == 1, 1], s=100, c="blue", label="Cluster 2")
plt.scatter(X2[y_hc2 == 2, 0], X2[y_hc2 == 2, 1], s=100, c="green", label="Cluster 3")
plt.xlabel("Average Credit Limit")
plt.ylabel("Total Credit Cards")
plt.title("Customer Clusters by Credit Limit and Card Count")
plt.legend()
plt.show()
```

## Step 4 - Evaluation Cluster Results using Silhouette Score

```python
# Measure clustering performance using the Silhouette Score
silhouette_2d = silhouette_score(X2, y_hc2)
print(f"Silhouette Score (using 2 features): {silhouette_2d:.3f}")
```
